Run this notebook on the 4090 machine. The notebook is only a launcher; the reproducible training pipeline lives in `scripts/run_ptbxl_linear_probe.sh`.

```python
from pathlib import Path

root = Path.cwd()
if not (root / "scripts" / "run_ptbxl_linear_probe.sh").exists():
    root = root.parent

assert (root / "scripts" / "run_ptbxl_linear_probe.sh").exists(), root
root
```

Default first run: PTB-XL `superdiagnostic`, full labels, HeartLang pretrained linear probing.

```python
import os
import subprocess

env = os.environ.copy()
env.update({
    "TASK": "superdiagnostic",
    "SPLIT_RATIOS": "1",
    "TRAINABLE": "linear",
    "EPOCHS": "100",
    "CUDA_VISIBLE_DEVICES": "0",
})

subprocess.run(
    ["bash", "scripts/run_ptbxl_linear_probe.sh"],
    cwd=root,
    env=env,
    check=True,
)
```
